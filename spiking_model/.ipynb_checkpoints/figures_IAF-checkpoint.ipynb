```python
import matplotlib.pyplot as plt
import numpy as np
import math as math
```

```python
timeTot = 10000; dt = 0.05; countBin = 200
nTime = math.ceil(timeTot/dt)

Ne = 4000; Ni = 1000; Na = 10
NeR = 200; NiR = 50; NaR = 10
percKicked = 0.1

fold_fig = 'figures/'
```

```python
# without astrocytes
save_fig = 1
folder = 'data/woastro/'

Lind_spikeA = np.loadtxt(folder+'Lind_spikeA.csv', delimiter=",")
Lind_spikeE = np.loadtxt(folder+'Lind_spikeE.csv', delimiter=",")
Lind_spikeI = np.loadtxt(folder+'Lind_spikeI.csv', delimiter=",")

rateA = np.loadtxt(folder+'rateA.csv', delimiter=",")
rateE = np.loadtxt(folder+'rateE.csv', delimiter=",")
rateI = np.loadtxt(folder+'rateI.csv', delimiter=",")
LIa = np.loadtxt(folder+'LIa.csv', delimiter=",")

LVa = np.loadtxt(folder+'LVa.csv', delimiter=",")
LVe = np.loadtxt(folder+'LVe.csv', delimiter=",")
LVi = np.loadtxt(folder+'LVi.csv', delimiter=",")

# Raster plot
plt.figure(figsize=(15,4))
idxE=np.where(Lind_spikeE[:]==1)
plt.scatter(idxE[1],idxE[0]+NiR,color='r')
idxI=np.where(Lind_spikeI[:,:]==1)
plt.scatter(idxI[1],idxI[0],color='b')
plt.xlabel('time (s)',fontsize=25)
plt.ylabel('cell',fontsize=25)
plt.xticks(np.array([0,200,400,600,800,1000])*200,[0,2,4,6,8,10],fontsize=25)
plt.yticks([0,50,100,150,200,250],fontsize=25)
if save_fig == 1:
    plt.savefig(fold_fig+'/raster_woastro.pdf',bbox_inches='tight')

# Voltage samples
fig, ax = plt.subplots(3,figsize=(15,4),sharex=True, gridspec_kw={'hspace': 0})
ax[0].plot(np.arange(nTime)[:]*dt/1e3,LVe[:],color='C3')
for i in range(len(idxE[1])):
    if idxE[0][i]==math.ceil(percKicked*Ne)+10-380:
        ax[0].axvline(idxE[1][i]*dt/1e3,ymin=0.4,ymax=1,color='C3')
ax[0].set_ylabel('V (mV)',fontsize=25)
ax[0].tick_params(axis='y', labelsize=25)
ax[0].set_yticks([0,15])
ax[0].set_ylim([0,30])

ax[1].plot(np.arange(nTime)[:]*dt/1e3,LVi[:],color='C0')
for i in range(len(idxI[1])):
    if idxI[0][i]==math.ceil(percKicked*Ni)+10-95:
        ax[1].axvline(idxI[1][i]*dt/1e3,ymin=0.4,ymax=1,color='C0')

plt.xticks(fontsize=25)
ax[1].set_yticks([0,15])
ax[1].tick_params(axis='y',labelsize=25)
ax[1].set_ylim([0,30])


ax[2].plot(np.arange(nTime)[:]*dt/1e3,LIa[:],'k')
ax[2].tick_params(axis='y',labelsize=25)
ax[2].set_ylim([0,0.004])
ax[2].set_yticks([0,2e-3])

plt.xlabel('time (s)',fontsize=25)
if save_fig == 1:
    plt.savefig(fold_fig+'/volt_woastro.pdf',bbox_inches='tight')

# Population rates
plt.figure(figsize=(15,4))
plt.plot(np.asarray(rateE)[:]/Ne*1000/(countBin*dt),'r',label='$r_E$')
plt.plot(np.asarray(rateI)[:]/Ni*1000/(countBin*dt),'b',label='$r_I$')
plt.xlabel('time (s)',fontsize=25)
plt.ylabel('Avg. rate $r_X$ (spks/s)',fontsize=25)
plt.xticks([0,200,400,600,800,1000],[0,2,4,6,8,10],fontsize=25)
plt.yticks([0,20,40,60,80,100],[0,20,40,60,80,100],fontsize=25)
plt.legend(fontsize=25)
if save_fig == 1:
    plt.savefig(fold_fig+'/rate_woastro.pdf',bbox_inches='tight')

```

```python
# with astrocytes, Jai>0
save_fig = 1
folder = 'data/Jai_sup_0/'

Lind_spikeA = np.loadtxt(folder+'Lind_spikeA.csv', delimiter=",")#read_csv('Lind_spikeA.csv')
Lind_spikeE = np.loadtxt(folder+'Lind_spikeE.csv', delimiter=",")#read_csv('Lind_spikeA.csv')
Lind_spikeI = np.loadtxt(folder+'Lind_spikeI.csv', delimiter=",")#read_csv('Lind_spikeA.csv')

rateA = np.loadtxt(folder+'rateA.csv', delimiter=",")#read_csv('Lind_spikeA.csv')
rateE = np.loadtxt(folder+'rateE.csv', delimiter=",")#read_csv('Lind_spikeA.csv')
rateI = np.loadtxt(folder+'rateI.csv', delimiter=",")#read_csv('Lind_spikeA.csv')
LIa = np.loadtxt(folder+'LIa.csv', delimiter=",")#read_csv('Lind_spikeA.csv')

LVa = np.loadtxt(folder+'LVa.csv', delimiter=",")#read_csv('Lind_spikeA.csv')
LVe = np.loadtxt(folder+'LVe.csv', delimiter=",")#read_csv('Lind_spikeA.csv')
LVi = np.loadtxt(folder+'LVi.csv', delimiter=",")#read_csv('Lind_spikeA.csv')

# Raster plot
plt.figure(figsize=(15,4))
idxA=np.where(Lind_spikeA[:,200:]==1)
plt.scatter(idxA[1],idxA[0]+NeR+NiR,color='g')
idxE=np.where(Lind_spikeE[:,200:]==1)
plt.scatter(idxE[1],idxE[0]+NiR,color='r')
idxI=np.where(Lind_spikeI[:,200:]==1)
plt.scatter(idxI[1],idxI[0],color='b')
plt.xlabel('time (s)',fontsize=25)
plt.ylabel('cell',fontsize=25)
plt.xticks(np.array([0,200,400,600,800,1000])*200,[0,2,4,6,8,10],fontsize=25)
plt.yticks(fontsize=25)
if save_fig == 1:
    plt.savefig(fold_fig+'/rasterJai_sup_0.pdf',bbox_inches='tight')

# Voltage samples
fig, ax = plt.subplots(4,figsize=(15,4),sharex=True, gridspec_kw={'hspace': 0})
ax[0].plot(np.arange(nTime)[200:]*dt/1e3,LVe[200:],color='C3')
for i in range(len(idxE[1])):
    if idxE[0][i]==math.ceil(percKicked*Ne)+10-380:
        ax[0].axvline(idxE[1][i]*dt/1e3,ymin=0.4,ymax=1,color='C3')
ax[0].set_ylabel('V (mV)',fontsize=25)
ax[0].tick_params(axis='y', labelsize=25)
ax[0].set_yticks([0,15])
ax[0].set_ylim([0,30])

ax[1].plot(np.arange(nTime)[200:]*dt/1e3,LVi[200:],color='C0')
for i in range(len(idxI[1])):
    if idxI[0][i]==math.ceil(percKicked*Ni)+10-95:
        ax[1].axvline(idxI[1][i]*dt/1e3,ymin=0.4,ymax=1,color='C0')

plt.xticks(fontsize=25)
ax[1].set_yticks([0,15])
ax[1].tick_params(axis='y',labelsize=25)
ax[1].set_ylim([0,30])


ax[2].plot(np.arange(nTime)[200:]*dt/1e3,LVa[200:],color='C2')
for i in range(len(idxA[1])):#Lind_spikeE[math.ceil(percKicked*Ne)+10])):
    if idxA[0][i]==0:
        ax[2].axvline(idxA[1][i]*dt/1e3,ymin=0.4,ymax=1,color='C2')
ax[2].set_yticks([0,15])
ax[2].tick_params(axis='y',labelsize=25)
ax[2].set_ylim([0,30])

ax[3].plot(np.arange(nTime)[200:]*dt/1e3,LIa[200:],'k')
ax[3].tick_params(axis='y',labelsize=25)
ax[3].set_ylim([0,0.004])
ax[3].set_yticks([0,2e-3])

plt.xlabel('time (s)',fontsize=25)
if save_fig == 1:
    plt.savefig(fold_fig+'/voltJai_sup_0.pdf',bbox_inches='tight')

# Population rates
plt.figure(figsize=(15,4))
plt.plot(np.asarray(rateE[1:])/Ne*1000/(countBin*dt),'r',label='$r_E$')
plt.plot(np.asarray(rateI[1:])/Ni*1000/(countBin*dt),'b',label='$r_I$')
plt.plot(np.asarray(rateA[1:])/Na*1000/(countBin*dt),'g',label='$r_A$')
plt.xlabel('time (s)',fontsize=25)
plt.ylabel('Avg. rate $r_X$ (spks/s)',fontsize=25)

plt.xticks([0,200,400,600,800,1000],[0,2,4,6,8,10],fontsize=25)
plt.yticks([0,20,40],[0,20,40],fontsize=25)
plt.legend(fontsize=25)
if save_fig == 1:
    plt.savefig(fold_fig+'/rateJai_sup_0.pdf',bbox_inches='tight')

```
